## Methods and Attributes between Series and DataFrames
- A **DataFrame** is a 2-dimensional table consisting of rows and columns.

- Like with a **Series**, Pandas assigns an index position/label to each **DataFrame** row.
- The **DataFrame** and **Series** have common and exclusive methods/attributes.
- The `hasnans` attribute exists only a **Series**. The `columns` attribute exists only on a **DataFrame**.
- Some methods/attributes will return different types of data.
- The `info` method returns a summary of the pandas object.

## Differences between Shared Methods

- On a **DataFrame**, the `sum` method defaults to adding the values by traversing the index (row values).
- The `axis` parameter customizes the direction that we add across. Pass `"columns"` or `1` to add "across" the columns.

```python
pd.read_csv('revenue.csv')
```

```python
revenue = pd.read_csv('revenue.csv' ,index_col=['Date'])
revenue
```

```python
s = pd.Series([1,2,3])
sum(s) #or 
s.sum()
```

```python
revenue.sum()
```

```python
revenue.sum(axis= 'index')  #calculates the entire values of each column 
```

```python
revenue.sum(axis='columns')  #gives the calculated value ,calculating row wise or  for each date total sales from 3 columns for specific date
```

## Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

```python
nba = pd.read_csv('nba.csv')
nba
```

```python
nba.dropna()  #by defualt it will remove any row , if it has a even a single missing value  , both are same syntax
nba.dropna(how='any')   #removes any row that has even a single missing value 
```

```python
nba.dropna(how='all')  #removes any row which has missing values , means the row which has missings value in each column , if theres a single value present 
                        # it will not remove that row
```

```python
nba.dropna(subset=['Salary'])  #with subset we can work with a column, if there is a missing a value in that column, it will remove that row
nba.dropna(subset=['Salary' ,'College']) #we can also add two columns, if either or both of them has any missing values , that row will be removed
```

## Fill in Missing Values with the fillna Method
- The `fillna` method replaces missing `NaN` values with its argument.
- The `fillna` method is available on both **DataFrames** and **Series**.
- An extracted **Series** is a view on the original **DataFrame**, but the `fillna` method returns a copy.

```python
nba = pd.read_csv('nba.csv').dropna(how='all')
nba
```

```python
nba.fillna(0)
```

```python
nba['Salary'] #this is a view of original dataframe 

nba['Salary'].fillna(0) #when we call fillna method, this returns a copy
```

```python
nba['Salary'] = nba['Salary'].fillna(0)
```

```python
nba
```

```python
nba['College'] = nba['College'].fillna('unknown')
```

```python
nba.head(17)
```

## The astype Method I
- The `astype` method converts a **Series's** values to a specified type.
- Pass in the specified type as either a string or the core Python data type.
- Pandas cannot convert `NaN` values to numeric types, so we need to eliminate/replace them before we perform the conversion.
- The `dtypes` attribute returns a **Series** with the **DataFrame's** columns and their types.

```python
nba = pd.read_csv('nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0)
nba['Weight'] = nba['Weight'].fillna(0) #filling missing values with 0
nba
```

```python
nba.dtypes
```

```python
nba['Salary'].astype(int)
nba['Salary'] = nba['Salary'].astype(int)  #now the salary column will have int values, dataframe is also affected
#now targeting weight column
nba['Weight'] = nba['Weight'].astype(int)

#after this, if we call dtype and check datatype of each column, it will show int for changes made above
```

```python
nba
```

```python
nba.dtypes
```

## The astype Method II
- The `category` type is ideal for columns with a limited number of unique values.
- The `nunique` method will return a **Series** with the number of unique values in each column.
- With categories, pandas does not create a separate value in memory for each "cell". Rather, the cells point to a single copy for each unique value.

```python
nba = pd.read_csv('nba.csv').dropna(how='all')
nba.tail()
```

```python
nba['Team'].nunique()  #retunns how many unique values there are
```

```python
nba.nunique()  #this way it will tell how many unique value are there in each column
```

```python
nba['Position'] =nba['Position'].astype('category')  #category is best to use when there is limited number or small number of unique values
```

```python
nba['Team'] = nba['Team'].astype('category')
```

```python
#now if we check info , and datatype it will show category
nba.info()
```

## Sort a DataFrame with the sort_values Method I
- The `sort_values` method sorts a **DataFrame** by the values in one or more columns. The default sort is an ascending one (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- If sorting by a single column, pass a string with its name.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.

```python
nba = pd.read_csv('nba.csv')
nba.tail()
```

```python
nba['Name'].sort_values()  #this way, we get a series then sort values
```

```python
#if want to sort the column in the data frame itself then
nba.sort_values('Name')   #the column values are sorted , by default it sorts in ascending order
```

```python
nba.sort_values('Name',ascending=False)  #this way it sorts the column in descending order
```

```python
nba.sort_values('Salary')  #in salary column there are missing values, by default, it will come at bottom
#but the values will be sorted in ascending or reverse order

nba.sort_values('Salary',ascending=False) #here values are sorted in descending order
```

```python
nba.sort_values('Salary',na_position='last')  #this parameter indicates, where the NaN missing values will be placed, first or last
nba.sort_values('Salary',na_position= 'first')
```

```python
#here we are sorting , salary column, values will be descending order, but missing values will be placed first
nba.sort_values('Salary',na_position= 'first' ,ascending=False) 
```

## Sort a DataFrame with the sort_values Method II
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

```python
nba = pd.read_csv('nba.csv')
nba.tail()
```

```python
nba.sort_values(['Team','Name'])
nba.sort_values(by=['Team','Name'], ascending=True) 
nba.sort_values(by=['Team','Name'], ascending=False)
```

```python
#to sort one column in ascending order, and other in descending order ,we can give True and false , which will be respectively aplied to column names
nba.sort_values(by=['Team','Name'], ascending=[True,False])  #here True is given to Team ,False is given to Name respectively
```

```python
nba.sort_values(['Position','Salary'], ascending=[True, False])
nba.sort_values(['Position', 'Salary'], ascending = False)
nba.sort_values(['Position', 'Salary'], ascending = [False,True])
```

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each **Series** value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

```python
nba = pd.read_csv('nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')#to salary colufilling missing values with 0, then passing as astype to replace floating type with integer type
nba                                                   #assigning this changes to salary column
```

```python
nba['Salary'].rank() #by defult values of salary column are assigned in ascending order , the highest value has been given the highes rank, lowest value has the lowest rank
```

```python
nba['Salary'].rank(ascending=False).astype(int) #now it is reversed, highest salary is given lowest num like 1 , 2 ,3 so on
#assigning it to the existing nba dataset, to a new column 
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype(int)
```

```python
#now to get the highest salaryt value , we can use sort_values method 

nba.sort_values('Salary Rank') #sorting this column in ascending order will give tbe lowest rank number, which is given to higest salary 

nba.sort_values('Salary', ascending=False).head(10) # we can use this way also , default it will sorted in ascending so have to give false for descending order for 
                                            # highest value to lowest 
```
