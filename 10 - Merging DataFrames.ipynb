## The pd.concat Function I
- The `concat` function concatenates one **DataFrame** to the end of another.
- The original index labels will be kept by default. Set `ignore_index` to True to generate a new index.
- The `keys` parameter create a **MultiIndex** using the specified keys/labels.

```python
pd.read_csv('restaurant_foods.csv')
pd.read_csv('restaurant_customers.csv')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
week2.head()
```

```python
len(week1)
```

```python
len(week2)
```

```python
pd.concat([week1 , week2])  #The concat function concatenates one DataFrame to the end of another.

pd.concat([week1 , week2], ignore_index = False)  #by default it is False,  keep the original index , do not create new one
pd.concat([week1 , week2], ignore_index = True)  #ignore dataframe indexes and create a new one 
```

```python
#if we want to keep original index of dataframe 
pd.concat([week1 , week2], ignore_index = False).sort_index() #here each one is going to occur more than once , then it is hard to figure out which dataframe did each row come from
                                                             #so this is not a optimized solution

# we use another parameter 'keys'
#the order should match the datframe we concatinate
pd.concat([week1 , week2], keys=['week 1' , 'week 2'])    #The keys parameter create a MultiIndex using the specified keys/labels.
```

## The pd.concat Function II
- Pandas will concatenate the **DataFrames** along the row/index axis.
- Pandas will include all columns that exist in either **DataFrame**. If there are no matching values, pandas will use `NaN` values.
- We can pass the `axis` parameter an argument of `"columns"` to concatenate on the column axis.

```python
df1 = pd.DataFrame([1,2,3] , columns=['A'])  #in [], we have values , and column name is A
df1
```

```python
df2 = pd.DataFrame([4,5,6] , columns=['B'])
df2
```

```python
pd.concat([df1 , df2])
pd.concat([df1, df2] , axis='index')
```

```python
#to match the value df2 df1 in same index , we pass axis = col , means adding df2 after df1 in col axis , besides df1 
pd.concat([df1 ,df2] , axis='columns')
```

```python
a = pd.DataFrame([1,2,3,4] , columns=['A'])
b = pd.DataFrame([1,2,4] , columns=['B'])
pd.concat([a , b] ,axis='columns')
```

```python
pd.concat([a ,b], keys=['a' , 'b'])   #with keys parameter we crete multiindex dataframe
```

## Left Joins
- The `merge` method joins two **DataFrames** together based on shared values in a column or an index.
- A left join merges one **DataFrame** into another based on values in the first one.
- The "left" **DataFrame** is the one we invoke the `merge` method on.
- If the left **DataFrame's** value is not found in the right **DataFrame**, the row will hold `NaN` values.
<img src="SQL_Joins.png" width="800" height="800"/>

```python
foods = pd.read_csv('restaurant_foods.csv')
customrers = pd.read_csv('restaurant_customers.csv')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
foods.head()
```

```python
week1.merge(foods , how='left' , on='Food ID')
```

## The left_on and right_on Parameters
- The `left_on` and `right_on` parameters designate the column names from each **DataFrame** to use in the merge.

```python
foods = pd.read_csv('restaurant_foods.csv')
customrers = pd.read_csv('restaurant_customers.csv')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week2.head()
```

```python
customrers.head()
```

```python
week2.merge(customrers , how='left' , left_on='Customer ID' , right_on='ID')  #The left_on and right_on parameters designate the column names from each DataFrame to use in the merge.
```

```python
#if we want to remove one column ,which has duplicate values , then we can use drop method to remove that column from the dataframe
#we have to mention axis column, pandas will look for that col in col axis
week2.merge(customrers, how='left' , left_on='Customer ID' , right_on='ID').drop('ID' , axis='columns')  .head()   
```

## Inner Joins I
- Inner joins merge two tables based on *shared*/*common* values in columns.
- If only one **DataFrame** has a value, pandas will exclude it from the final results set.
- If the same ID occurs multiple times, pandas will store each possible combination of the values.
- The design of the join ensures that the results will be the same no matter what **DataFrame** the `merge` method is invoked upon.
<img src="SQL_Joins.png" width="800" height="800"/>

```python
foods = pd.read_csv('restaurant_foods.csv')
customers = pd.read_csv('restaurant_customers.csv')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
week2.head()
```

```python
week1.merge(week2 , how='inner' , on='Customer ID')
```

```python
#if we want to add names at the end of the col name instead of '_x' or '_y' , we can use suffixes ,with it we can customize the name of the columns 
week1.merge(week2 , how='inner' , on='Customer ID' , suffixes=['- Week1' , '- Week2'])
```

```python
week1[week1['Customer ID'] == 537]
```

```python
week2[week2['Customer ID'] == 537]
```

```python
week1[week1['Customer ID'] == 155]
```

```python
week2[week2['Customer ID'] == 155]
```

## Full/Outer Join
- A **full/outer** joins values that are found in either **DataFrame** or both **DataFrames**.
- Pandas does not mind if a value exists in one **DataFrame** but not the other.
- If a value does not exist in one **DataFrame**, it will have a `NaN`.

<img src="SQL_Joins.png" width="800" height="800"/>

```python
foods = pd.read_csv('restaurant_foods.csv')
customers = pd.read_csv('restaurant_customers.csv')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
week2.head()
```

```python
week1.merge(week2 , how='outer' , on='Customer ID')
week1.merge(week2 , how='outer' , on='Customer ID' , suffixes=[' - week1' , ' - week2'])

#in row 1, for week1 we have value 4 for cus.id 97 , but we do nat have value for week2 for cus.id 97 so it has NaN 
#similarly if we have values present for cus.ID in one dataframe it will show, but if we do not have value for same cus.id for other datframe it will show NaN
```

```python
week1.merge(week2 , how='outer' , on='Customer ID' , suffixes=[' - week1' , ' - week2'] , indicator=True) #this parameter indicates whether the value is 
                                                                                                         #present left dataframe , right or both dataframe
```

```python
merged = week1.merge(week2 , how='outer' , on='Customer ID' , suffixes=[' - week1' , ' - week2'] , indicator=True)
merged['_merge']
```

```python
merged['_merge'].value_counts()
```

```python
#want capture values exist in only one table , do not care if its in left or right table , but exist in only one table 
merged[merged['_merge'].isin(['left_only' , 'right_only'])]
```

## Merging by Indexes with the left_index and right_index Parameters
- Use the `on` parameter if the column(s) to be matched on have the same names in both **DataFrames**.
- Use the `left_on` and `right_on` parameters if the column(s) to be matched on have different names in the two **DataFrames**.
- Use the `left_index` or `right_index` parameters (set to True) to if the values to be matched on are found in the index of a **DataFrame**.

```python
foods = pd.read_csv('restaurant_foods.csv', index_col='Food ID')
customers = pd.read_csv('restaurant_customers.csv', index_col= 'ID')
week1 = pd.read_csv('restaurant_week_1_sales.csv')
week2 = pd.read_csv('restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
customers.head()
```

```python
foods.head()
```

```python
#using left join on week1 and customers table

#left dataframe week1 use the customerID col , and right datframe customers , use the index 

week1_with_customer = week1.merge(customers , how='left' , left_on='Customer ID' , right_index=True) 
week1_with_customer
```

```python
#we continue usinf merging with foods table

week1.merge(
    customers , how='left' , left_on='Customer ID' , right_index=True
).merge(foods , how='left' , left_on='Food ID' , right_index=True)

#----------------------------

second_merge_with_foods =  week1_with_customer.merge(foods , how='left' , left_on='Food ID' , right_index=True)
second_merge_with_foods
```
